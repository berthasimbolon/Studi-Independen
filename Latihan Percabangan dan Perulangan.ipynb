```python
# Percabangan dan Perulangan Luas dan Keliling Lingkaran
print("MENGHITUNG LUAS & KELILING lINGKARAN")

def hitung():
  jari_jari = float(input("\nMasukan jari-jari dalam cm: "))
  print("\nPilih 1 untuk hitung luas\nPilih 2 untuk hitung keliling")
  pilihan_kalkulasi = input("\nInput pilihan:")
  phi = 3.14

  if pilihan_kalkulasi == "1":
    luas_lingkaran = phi * jari_jari**2
    print("\nLuas Lingkaran adalah:", luas_lingkaran, "cm^2")
  elif pilihan_kalkulasi == "2":
    keliling_lingkaran = 2 * phi * jari_jari
    print("\nKeliling Lingkaran adalah:", keliling_lingkaran, "cm")
  else:
    print("\nMaaf, pilihan tidak tersedia")


if __name__ == '__main__':
  while True:
    print("\nPilih 1 untuk melanjutkan\nPilih 2 untuk keluar")
    pilihan_IO = input("\nInput pilihan: ")

    if pilihan_IO == "1":
      hitung()
    elif pilihan_IO == "2":
      break
    else:
      print("Pilihan tidak tersedia")

  print("\nTerima kasih telah menggunakan program ini.")
```

```python
#Percabangan dan Perulangan Volume dan Luas Permukaan Balok
print("MENGHITUNG VOLUME & LUAS PERMUKAAN BALOK")

def hitung():
  p = float(input("\nMasukan Panjang Sisi dalam cm: "))
  l = float(input("\nMasukan Lebar Sisi dalam cm: "))
  t = float(input("\nMasukan Tinggi Sisi dalam cm: "))
  print("Pilih 1 untuk menghitung volume\nPilih 2 untuk menhitung luas permukaan")
  pilihan_kalkulasi = input("\nInput pilihan:")
  
  if pilihan_kalkulasi == "1":
    volume = p * l * t
    print("\nVolume Balok:", volume, "cm^3")
  elif pilihan_kalkulasi == "2":
    luas_permukaan = 2 * (p*l + p*t + l*t)
    print("\nLuas Permukaan adalah:", luas_permukaan, "cm^2")
  else:
    print("\nMaaf, pilihan tidak tersedia")

if __name__ == '__main__':
  while True:
    print("\nPilih 1 untuk melanjutkan\nPilih 2 untuk keluar")
    pilihan_IO = input("\nInput pilihan")

    if pilihan_IO == "1":
      hitung()
    elif pilihan_IO == "2":
      break
    else:
      print("Pilihan tidak tersedia")


  print("\nTerima kasih telah menggunakan program ini.")
```
